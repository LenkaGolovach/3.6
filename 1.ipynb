### У нас есть функция двух переменных $g(x, y)$, которая представляет собой высоту точки на горной местности в зависимости от её координат $x$ и $y$. Наша цель - найти самую высокую точку на горной местности.

### Импортируем библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Определение функции высоты
def g(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
```

### Создание сетки значений x и y

```python
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
```

```python
# Вычисление значений высоты на сетке
Z = g(X, Y)
```

### Нахождение координат самой высокой точки

```python
max_height = np.max(Z)
max_index = np.unravel_index(np.argmax(Z), Z.shape)
max_x = X[max_index]
max_y = Y[max_index]
```

### Построение трехмерного графика

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='terrain')

# Настройка осей
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Height')

plt.title('Горная местность')
plt.show()
# Вывод результатов
print(f"Самая высокая точка: высота={max_height}, координаты: x={max_x}, y={max_y}")
```
